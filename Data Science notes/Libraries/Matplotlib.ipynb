```python
import matplotlib.pyplot as plt
import matplotlib as mp
import numpy as np
```

1.1 Generating Sin and Cos Graph using Matplotlib

```python
x=np.linspace(0,4*np.pi,100)
sin_y=np.sin(x)
cos_y=np.cos(x)
print(sin_y)
print(cos_y)
plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cosine')
plt.legend(loc = 'upper right')
plt.title('Sine Wave')
```

```python
x=np.linspace(0,3*np.pi,100)
sin_x=np.sin(x)
cos_x=np.cos(x)
plt.plot(x,sin_x,'*',label='sine')
plt.plot(x,cos_x,'--',label='cosine')
plt.title('Sine Wave')
plt.show()
```

```python
plt.style.use('seaborn-dark')
plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cosine')
plt.legend(loc = 'upper right')
plt.title('Sine Wave')
```

```python
plt.style.use('grayscale')
plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cosine')
plt.legend(loc = 'upper right')
plt.title('Sine Wave')
```
